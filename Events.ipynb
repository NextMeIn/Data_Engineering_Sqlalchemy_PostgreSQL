```python
import json
import requests
import io
import datetime
import pandas as pd
import numpy as np
import databaseconfig as dbconnect
```

```python
# events
url = "https://opendata.tpl.ca/resources/events"
response = requests.get(url)
events = response.json()
len(events)
```

```python
events[100]
```

```python
dt = datetime.date.today().strftime("%Y/%m/%d")
dt
```

```python
future_events = [event for event in events if event['date1'] > dt]
```

```python
print(len(future_events))
future_events[18]
```

```python
future_events[0].keys()
```

```python
future_events_df = pd.DataFrame(future_events)
```

```python
future_events_df.head()
future_events_df = future_events_df[['title','date','time','enddate','description','library','link','date1','agegroups']]
```

```python
future_events_df
```

```python
ID = list(np.arange(len(future_events_df))+1)
len(ID)
```

```python
future_events_df['id'] = ID
```

```python
future_events_df.head()
```

```python
future_events_df.keys()
```

```python
future_events_df = future_events_df.rename(columns={'agegroups':'age_group'})
```

```python
from sqlalchemy import create_engine
rds_connection_string = 'postgres:'+dbconnect.postgres['passwd']+'@localhost:5432/toronto_library_db'
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
tables = list(engine.table_names())
tables.sort()
tables
```

```python
tables[7]
```

```python
future_events_df.to_sql(name=tables[7],con=engine,if_exists='append',index=False)
```
