```python
from splinter import Browser
from bs4 import BeautifulSoup
```

# Mac Users

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

# Windows Users

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.torontopubliclibrary.ca/books-video-music/new-items.jsp?category=Adult+Fiction'
browser.visit(url)
```

```python
# Iterate through all pages
for x in range(50):
    # HTML object
    html = browser.html
    # Parse HTML with Beautiful Soup
    soup = BeautifulSoup(html, 'html.parser')
    # Retrieve all elements that contain book information
    #titles = soup.find_all('span', class_='ellipsis_text')
    
    details = soup.find_all('div', class_='details')
    

    # Iterate through each book
    for book_details in details:
        # Use Beautiful Soup's find() method to navigate and retrieve attributes
        href = book_details.find('a')['href']
        title = book_details.find('span').text
        print('-----------')
        print(title)
        print('https://www.torontopubliclibrary.ca/books-video-music/new-items.jsp?category=Adult+Fiction' + href)

    # Click the 'Next' button on each page
    try:
        browser.click_link_by_partial_text('Next')
          
    except:
        print("Scraping Complete")
```
